```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = yf.download(
    tickers = [
        '^GSPE', # Energy
        '^SP500-15', # Materials
        '^SP500-20', # Industrials
        '^SP500-25', # Consumer Discretionary
        '^SP500-30', # Consumer Staples
        '^SP500-35', # Health Care
        '^SP500-40', # Financials
        '^SP500-45', # Information Technology
        '^SP500-50', # Communication Services
        '^SP500-55', # Utilities
        '^SP500-60', # Real Estate
    ],
    start = '1980-01-01',
    progress = False,
    multi_level_index = True,
    auto_adjust = False
)
```

```python
data_new = data[['Open','High','Low','Close','Volume']].copy(
    'deep'
)
```

```python
data_new.xs('^GSPE', axis = 1, level = 1)
```

```python
asset_tickers = [
        '^GSPE', # Energy
        '^SP500-15', # Materials
        '^SP500-20', # Industrials
        '^SP500-25', # Consumer Discretionary
        '^SP500-30', # Consumer Staples
        '^SP500-35', # Health Care
        '^SP500-40', # Financials
        '^SP500-45', # Information Technology
        '^SP500-50', # Communication Services
        '^SP500-55', # Utilities
        '^SP500-60', # Real Estate
    ]

mapping_dict = {
    '^GSPE' : 'Energy',
    '^SP500-15' : 'Materials',
    '^SP500-20' : 'Industrials',
    '^SP500-25' : 'Consumer Discretionary',
    '^SP500-30' : 'Consumer Staples',
    '^SP500-35' : 'Health Care',
    '^SP500-40' : 'Financials',
    '^SP500-45' : 'Information Technology',
    '^SP500-50' : 'Communication Services',
    '^SP500-55' : 'Utilities',
    '^SP500-60' : 'Real Estate',
}
```

```python
with pd.ExcelWriter('../../data/market/SP500_sectors.xlsx') as writer:
    for ticker in asset_tickers:
        data_new.xs(ticker, axis = 1, level = 1).to_excel(writer, sheet_name = mapping_dict[ticker])
```
