```python
import numpy as np
import pandas as pd

from src.dataloader.data_downloader import get_data_from_tradingview
```

```python
from tvDatafeed import Interval

data = get_data_from_tradingview(
    tickers = ['IBOV','SENSEX'],
    exchange = ['BMFBOVESPA','BSE'],
    interval = Interval.in_daily,
    n_bars = 13000,
    tz_cleansing = True,
    multi_level_index=False
)
```

```python
with pd.ExcelWriter('Foreign_index.xlsx') as writer :
    data.xs('IBOV', level = 1, axis= 1).dropna().to_excel(
        writer, sheet_name = 'IBOVESPA'
    )
    data.xs('SENSEX', axis = 1, level = 1).dropna().to_excel(
        writer, sheet_name = 'SENSEX'
    )
```
